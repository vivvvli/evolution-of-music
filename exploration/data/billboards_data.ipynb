```python
from urllib.request import urlopen
from bs4 import BeautifulSoup, Comment
import pandas as pd
import requests
import numpy as np
import os
from collections import OrderedDict
import main
sp = main.sp

def append_song_stats(date):
    chart = billboard.ChartData('hot-100', date)
    song_list = []
    for idx,song in enumerate(chart):
        track_info=[]
        song_ranking = str(int(idx) + 1)
        song_name = song.title
        song_artist = song.artist
        track_info.append(date)
        track_info.append(song_ranking)
        track_info.append(song_name)
        track_info.append(song_artist)
        track_info.append(song.peakPos)
        track_info.append(song.lastPos)
        track_info.append(song.weeks)
        try:
            song_features = list(get_spotify_features(song_name,song_artist))
            for each in song_features:
                track_info.append(each)
        except: 
            song_features = 'N/A'
            for each in (0,18):
                track_info.append(song_features)       
        song_list.append(track_info)
    return song_list
```

```python
headers = ["date", 
        "song_ranking", 
        "song_name",
        "song_artist", 
        "peakPos", 
        "lastPos", 
        "weeks", 
        'danceability',
        'energy', 
        'key', 
        'loudness',
        'mode',
        'speechiness',
        'acousticness',
           'instrumentalnes',
           'liveness',
           'valence',
           'tempo',
           'type',
           'id',
           'uri',
           'track_href',
           'analysis_url',
           'duration_ms',
          'time_signature']
final_df = pd.DataFrame()

for year in range(2000,2021):
    for month in range(1,13):
        year_month = str(year) + '-' + str(month).zfill(2)
        df = pd.DataFrame(append_song_stats(year_month + '-01'))
        final_df = final_df.append(df)
final_df.columns = headers
```
