```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

xls = pd.ExcelFile('/Users/vivianli/Documents/Data Science/evolution-of-music/exploration/data/spotify_data/spotify_charts_features.xlsx')

map = {}
for sheet_name in xls.sheet_names:
    map[sheet_name] = xls.parse(sheet_name)
```

```python
# putting all the years together
composite_table = pd.DataFrame()
for key, value in map.items():
    temp1 = pd.DataFrame()
    temp1 = temp1.append(value)
    temp1['date'] = pd.to_datetime(key[-7:])
    composite_table = composite_table.append(temp1)
```
